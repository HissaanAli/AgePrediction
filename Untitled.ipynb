```python
TRAIN_PATH = "dataset/training_set/"
VAL_PATH = "dataset/validation_set/"
TEST_PATH = "dataset/test_set/"

def imshow(img):
    plt.imshow(img, cmap = 'gray')
def gray_to_rgb(img):
    return tf.image.grayscale_to_rgb(tf.convert_to_tensor(img.reshape(91,91,1)))
def resize(img):
    return tf.image.resize(img, [224,224], method='nearest').numpy()
```

# DATA LOADER

```python
def load_into_df(path):
    data = {
        'path': [],
        'age': []
           }
    file_names = os.listdir(path)
    for file in file_names:
        data['age'].append(int(file.split('/')[-1].split('_')[0]))
        data['path'].append(file)
    df = pd.DataFrame(data)
    return df
class Item:
    def __init__(self, path):
        self._image = np.array(Image.open(path))
        self._age = int(path.split('/')[-1].split('_')[0])
    def get_image(self):
        return self._image
    def get_age(self):
        return self._age
    
```

```python
train_df = load_into_df(TRAIN_PATH)
print('train_df:')
display(train_df.describe())
val_df = load_into_df(VAL_PATH)
print('val_df:')
display(val_df.describe())
test_df = load_into_df(TEST_PATH)
print('test_df:')
display(test_df.describe())
```

# DATA PREPARATION

```python
def get_xy(path):
    file_names = os.listdir(path)
    items = []
    for file in file_names:
        item = Item(path+file)
        items.append(item)
    #x = np.array([resize(gray_to_rgb(item.get_image())) for item in items])
    x = np.array([item.get_image().flatten() for item in items])
    y = np.array([item.get_age() for item in items])
    return x, y
```

```python
x_train, y_train = get_xy(TRAIN_PATH)
x_val, y_val = get_xy(VAL_PATH)
x_test, y_test = get_xy(TEST_PATH)

print('training:',x_train.shape, y_train.shape)
print('validation:',x_val.shape, y_val.shape)
print('test:',x_test.shape, y_test.shape)
```

# PRE PROCESSING

```python
# NORMALIZATION

def normalize(x):
    return x/255
x_train = normalize(x_train)
x_val = normalize(x_val)
x_test = normalize(x_test)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42, k_neighbors = 4)
x_train_new, y_train_new = sm.fit_resample(x_train, y_train)
x_train_new = np.array([x.reshape(-1,1).reshape(91,91, 1) for x in x_train_new])
```

```python
x_val = x_val.reshape(x_val.shape[0], 91, 91,1)
x_test = x_test.reshape(x_test.shape[0], 91, 91,1)
print(x_val.shape)
print(x_test.shape)
```

```python
res_model = ResNet50(include_top=False, weights='imagenet', input_shape=(224, 224, 3))#, input_tensor=Input(shape=(224, 224, 3)))

for layer in res_model.layers[:]:
    layer.trainable = False

model = res_model.output
#model = Conv2D(256, (7,7), padding = 'same', activation = 'relu')(model)
model = AveragePooling2D(pool_size = (3,3))(model)
model = BatchNormalization(name="batch_norm_1")(model)
model = Flatten(name="flatten")(model)
model = Dense(512, activation="relu")(model)
model = Dropout(0.3)(model)
model = BatchNormalization(name="batch_norm_2")(model)
#model = Dense(256, activation="relu")(model)
#model = Dropout(0.3)(model)
#model = BatchNormalization(name="batch_norm_3")(model)
#model = Dense(128, activation="relu")(model)
#model = Dropout(0.4)(model)
#model = BatchNormalization(name="batch_norm_4")(model)
#model = Dense(64, activation="relu")(model)
#model = Dropout(0.4)(model)
model = Dense(1, activation="linear")(model)


model = Model(inputs=res_model.input, outputs=model)
model.summary()
```

```python
model.compile(loss='mean_absolute_error',
              optimizer=tf.keras.optimizers.Adam(), 
              metrics=[tf.keras.metrics.MeanAbsoluteError()]
             )
```

```python
history = model.fit(x_train, y_train, validation_data = (x_val, y_val), epochs = 100)
```

```python
plt.rcParams["figure.figsize"] = (10,10)
loss = history.history["loss"]
val_loss = history.history["val_loss"]
plt.plot(loss, label = 'loss')
plt.plot(val_loss, label = 'val_loss')
```

```python
from keras.models import Model
from keras.layers.normalization import BatchNormalization
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.core import Activation
from keras.layers.core import Dropout
from keras.layers.core import Lambda
from keras.layers.core import Dense
from keras.layers import Flatten
from keras.layers import Input
import tensorflow as tf

class UtkMultiOutputModel():
    """
    Used to generate our multi-output model. This CNN contains three branches, one for age, other for 
    sex and another for race. Each branch contains a sequence of Convolutional Layers that is defined
    on the make_default_hidden_layers method.
    """
    def make_default_hidden_layers(self, inputs):
        """
        Used to generate a default set of hidden layers. The structure used in this network is defined as:
        
        Conv2D -> BatchNormalization -> Pooling -> Dropout
        """
        x = Conv2D(32, (3, 3), padding="same")(inputs)
        x = Activation("relu")(x)
        x = BatchNormalization(axis=-1)(x)
        x = MaxPooling2D(pool_size=(3, 3))(x)
        x = Dropout(0.25)(x)

        x = Conv2D(64, (3, 3), padding="same")(x)
        x = Activation("relu")(x)
        x = BatchNormalization(axis=-1)(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        x = Dropout(0.25)(x)

        x = Conv2D(64, (3, 3), padding="same")(x)
        x = Activation("relu")(x)
        x = BatchNormalization(axis=-1)(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        x = Dropout(0.3)(x)
        
        x = Conv2D(128, (3, 3), padding="same")(x)
        x = Activation("relu")(x)
        x = BatchNormalization(axis=-1)(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        x = Dropout(0.3)(x)
        
        x = Conv2D(128, (3, 3), padding="same")(x)
        x = Activation("relu")(x)
        x = BatchNormalization(axis=-1)(x)
        x = MaxPooling2D(pool_size=(2, 2))(x)
        x = Dropout(0.4)(x)
        
        
        #x = Conv2D(128, (3, 3), padding="same")(x)
        #x = Activation("relu")(x)
        #x = BatchNormalization(axis=-1)(x)
        #x = MaxPooling2D(pool_size=(2, 2))(x)
        #x = Dropout(0.35)(x)
        return x


    def build_age_branch(self, inputs):   
        """
        Used to build the age branch of our face recognition network.
        This branch is composed of three Conv -> BN -> Pool -> Dropout blocks, 
        followed by the Dense output layer.

        """
        x = self.make_default_hidden_layers(inputs)

        x = Flatten()(x)
        x = Dense(256)(x)
        x = Activation("relu")(x)
        x = BatchNormalization()(x)
        x = Dropout(0.5)(x)
        x = Dense(1)(x)
        x = Activation("linear", name="age_output")(x)

        return x

    def assemble_full_model(self, width, height):
        """
        Used to assemble our multi-output model CNN.
        """
        input_shape = (height, width, 1)

        inputs = Input(shape=input_shape)

        age_branch = self.build_age_branch(inputs)
        

        model = Model(inputs=inputs,
                     outputs = [age_branch],
                     name="face_net")

        return model
mod = UtkMultiOutputModel().assemble_full_model(91, 91)
```

```python
from keras.optimizers import Adam

init_lr = 1.2e-4
epochs = 400

opt = tf.keras.optimizers.Adam(lr=init_lr, decay=init_lr / epochs)

mod.compile(loss='mean_absolute_error',
              optimizer=opt, 
              metrics=[tf.keras.metrics.MeanAbsoluteError()]
             )
mod.summary()
```

```python
train_generator = train_datagen.flow(
        x_train,
        y_train,
        batch_size=16)
```
